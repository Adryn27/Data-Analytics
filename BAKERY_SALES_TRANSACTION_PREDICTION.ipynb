```python
#import library
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#Membaca data
data = pd.read_csv('bread basket.csv')
data.info()
```

```python
#Mengecek nilai kosong
print(data.isnull().sum())
```

Setelah dilakukan pengecekan, dataset tidak memiliki data yang bernilai null disetiap kolomnya, sehingga dataset aman digunakan walaupun tanpa pembersihan data terlebih dahulu

```python
#Menampilkan data
data
```

```python
#Melihat duplikasi data
data.nunique()
```

Setelah melihat duplikasi data pada masing-masing kolom, terlihat bahwa kolom weekday_weekend hanya memiliki 2 data saja. Sehingga, dilakukan grouping data pada masing-masing nilai yakni weekday dan weekend untuk tujuan pencarian pola-pola pembelian berdasarkan waktu tersebut untuk membentuk penataan produk dan kombinasi paket yang cocok di kedua waktu tersebut.

```python
#Grouping Weekday
group_weekday=(data[data['weekday_weekend']=='weekday']
       .groupby(['Transaction','Item'])['Item']
       .count().unstack().reset_index().fillna(0)
       .set_index('Transaction'))
group_weekday.head()
```

```python
#Grouping Weekend
group_weekend=(data[data['weekday_weekend']=='weekend']
       .groupby(['Transaction','Item'])['Item']
       .count().unstack().reset_index().fillna(0)
       .set_index('Transaction'))
group_weekend.head()
```

Sebelum ke tahap pemodelan, dilakukan encoding data guna untuk menunjukkan bahwa jika kuantitas nya kurang dari 0 maka tetap ditulis dengan 0, jika kuantitas nya lebih dari 1 maka tetap akan ditulis 1. Ini dilakukan karena pada association rule yang dihitung adalah kehadiran dari item tersebut, bukan dari kuantitas.

```python
#Encoding data
def encode_qty(x):
  if x <= 0:
    return 0
  if x >= 1:
    return 1
```

```python
#Encoding data grup weekday
encode_group_weekday=group_weekday.applymap(encode_qty)
encode_group_weekday.head()
```

```python
#Encoding data grup weekend
encode_group_weekend=group_weekend.applymap(encode_qty)
encode_group_weekend.head()
```

Untuk mempersempit ruang lingkup dan melihat pola pembelian pelanggan yakni berupa apa saja item atau produk yang sering dibeli baik pada waktu weekday dan weekend dengan menetapkan minimal support 4% atau item yang sering muncul dari 4% atau lebih jumlah transaksi pada dataset.

```python
#Mengelompokkan item/produk yang sering dibeli pada weekday
frequent_itemsets_day=apriori(encode_group_weekday, min_support=0.04, use_colnames=True)
frequent_itemsets_day.sort_values('support', ascending=False)
```
